Find `XMAS` in the lines, it can go in any direction (horizontal, vertical, diagonal, backwards). It cannot wrap around the edges.

````python
# file is relative to where the jupyter server is started
with open('days/4/input.txt', 'r') as f:
    data = f.read().splitlines()
    print(data)
````

````python
grid = [ list(line) for line in data ]

directions = {
    'right': (0,1),   # right
    'left': (0,-1),  # left
    'down': (1,0),   # down
    'up': (-1,0),  # up
    'diag_dr': (1,1),   # diagonal down-right
    'diag_ul': (-1,-1), # diagonal up-left
    'diag_ur': (-1,1),  # diagonal up-right
    'diag_dl': (1,-1)   # diagonal down-left
}

def is_valid_direction(y, x):
    if 0 <= x < len(grid[0]) and 0 <= y < len(grid):
        return True
    else:
        return False
````

````python
xmas_found = 0
for iy, y in enumerate(grid):
    for ix, x in enumerate(y):
        if x == 'X':
            for dir in directions:
                xchange = directions[dir][0]
                ychange = directions[dir][1]
                try:
                    my = iy + ychange
                    mx = ix + xchange
                    if is_valid_direction(my, mx) and grid[my][mx] == 'M':
                        ay = my + ychange
                        ax = mx + xchange
                        if is_valid_direction(ay, ax) and grid[ay][ax] == 'A':
                            sy = ay + ychange
                            sx = ax + xchange
                            if is_valid_direction(sy, sx) and grid[sy][sx] == 'S':
                                xmas_found += 1
                except IndexError:
                    continue 
print(xmas_found)
````

Looking for the instructions, you flip over the word search to find that this isn't actually an `_XMAS_` puzzle; it's an `_X-MAS_` puzzle in which you're supposed to find two `MAS` in the shape of an `X`. (`A` being the center of the X)

````python
# Only concerned with diagonals
x_mas_found = 0

for iy, y in enumerate(grid):
    for ix, x in enumerate(y):
        # if letter is `M` and ix + 2 is `S` and iy + 2 is `M` and (ix + 2, iy + 2) is `S` and (ix + 1, iy + 1) is `A`
        # if letter is `M` and ix + 2 is `M` and iy + 2 is `S` and (ix + 2, iy + 2) is `S` and (ix + 1, iy + 1) is `A`
        # if letter is `S` and ix + 2 is `M` and iy + 2 is `S` and (ix + 2, iy + 2) is `M` and (ix + 1, iy + 1) is `A`
        # if letter is `S` and ix + 2 is `S` and iy + 2 is `M` and (ix + 2, iy + 2) is `M` and (ix + 1, iy + 1) is `A`
        a_correct = is_valid_direction(iy + 1, ix + 1) and grid[iy + 1][ix + 1] == 'A'
        x_correct = is_valid_direction(iy, ix + 2)
        y_correct = is_valid_direction(iy + 2, ix)
        # This seems like a decent shortcut up front
        if a_correct and x_correct and y_correct:
            # And this seems probably overdone and naive
            m_to_m_on_x = x == 'M' and grid[iy][ix + 2] == 'M'
            m_to_m_on_y = x == 'M' and grid[iy + 2][ix] == 'M'
            m_to_s_on_x = x == 'M' and grid[iy][ix + 2] == 'S'
            m_to_s_on_y = x == 'M' and grid[iy + 2][ix] == 'S'

            s_to_s_on_x = x == 'S' and grid[iy][ix + 2] == 'S'
            s_to_s_on_y = x == 'S' and grid[iy + 2][ix] == 'S'
            s_to_m_on_x = x == 'S' and grid[iy][ix + 2] == 'M'
            s_to_m_on_y = x == 'S' and grid[iy + 2][ix] == 'M'
            
            m_to_s_on_xy = x == 'M' and grid[iy + 2][ix + 2] == 'S'
            s_to_m_on_xy = x == 'S' and grid[iy + 2][ix + 2] == 'M'
            
            if (
                (m_to_s_on_xy and m_to_m_on_x and m_to_s_on_y) or
                (m_to_s_on_xy and m_to_s_on_x and m_to_m_on_y) or
                (s_to_m_on_xy and s_to_s_on_x and s_to_m_on_y) or
                (s_to_m_on_xy and s_to_m_on_x and s_to_s_on_y)
            ):
                x_mas_found += 1


print(x_mas_found)
````
