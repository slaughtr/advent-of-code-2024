Scan the corrupted memory for uncorrupted mul instructions. What do you get if you add up all of the results of the multiplications?
_just_ looking for `mul(num,num)`

```python
# file is relative to where the jupyter server is started
with open('days/3/input.txt', 'r') as f:
    raw = f.read()
    print(raw)
```

```python
import re

maybe = raw.split('mul(')
total = 0

for part in maybe:
    matched = re.match(r'\d+,\d+\)', part)
    if matched is not None:
        matched_str = str(matched.group(0)).removesuffix(')')
        first, second = [ int(x) for x in matched_str.split(',') ]
        total += (first * second)

print(total)
```

Part two
There are two new instructions you'll need to handle:

- The `do()` instruction _enables_ future `mul` instructions.
- The `don't()` instruction _disables_ future `mul` instructions.

Only the _most recent_ `do()` or `don't()` instruction applies. At the beginning of the program, `mul` instructions are _enabled_.

```python

maybe = raw.split('mul(')
new_total = 0
enabled = True

for part in maybe:
    # We start with our mul() args so whatever disabled is at the time should be respected
    matched = re.match(r'\d+,\d+\)', part)
    if matched is not None and enabled == True:
        matched_str = str(matched.group(0)).removesuffix(')')
        first, second = [ int(x) for x in matched_str.split(',') ]
        new_total += (first * second)

    # find returns the index, or -1 if not found
    dont = part.find('don\'t()')
    do = part.find('do()')
    if dont != -1 and do != -1:
        print(f'Part had do and dont {part}')
        # do came after don't
        if dont < do:
            enabled = True
        # don't came after do
        elif dont > do:
            enabled = False
    elif dont != -1:
        enabled = False
    elif do != -1:
        enabled = True
    else:
        print(f'Part did not have do or dont {part}')

print(new_total)
```
